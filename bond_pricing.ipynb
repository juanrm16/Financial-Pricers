Bond pricing model in Python

1. Pricing Functions

```python
def bond_price(face_value, coupon_rate, yield_to_maturity, years, freq=2):
    coupon = face_value * coupon_rate / freq
    n = int(years * freq)
    pv = 0
    for t in range(1, n + 1):
        pv += coupon / (1 + yield_to_maturity / freq) ** t
    pv += face_value / (1 + yield_to_maturity / freq) ** n
    return pv
```

2. Input Some Example Parameters & Calculate Price

```python
if __name__ == "__main__":
    price = bond_price(1000, 0.05, 0.04, 5, 2)
    print("Bond Price:", price)
```

3. Visualize

```python
import numpy as np
import matplotlib.pyplot as plt

face_value = 1000         
coupon_rate = 0.05       
years = 10          
freq = 1

def bond_price(face, coupon, ytm, periods, freq=1):
    coupon_payment = face * coupon / freq
    return sum([coupon_payment / (1 + ytm / freq) ** (freq * (i + 1) / freq) for i in range(periods)]) + face / (1 + ytm / freq) ** periods

ytm_values = np.linspace(0.01, 0.10, 100)
bond_prices = [bond_price(face_value, coupon_rate, ytm, years, freq) for ytm in ytm_values]

plt.figure(figsize=(7,4))
plt.plot(ytm_values * 100, bond_prices)
plt.xlabel('Yield to Maturity (%)')
plt.ylabel('Bond Price ($)')
plt.title('Bond Price vs. Yield to Maturity')
plt.grid(True)
plt.tight_layout()
plt.show()
```
