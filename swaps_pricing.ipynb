Swaps pricing model in Python

1. Define the Fair Fixed Rate Function

```python
def fair_fixed_rate(discount_factors, accruals):
    annuity = sum(a * df for a, df in zip(accruals, discount_factors))
    df_last = discount_factors[-1]
    return (1 - df_last) / annuity
```

2. Input Some Example Parameters

```python
dfs = [0.995, 0.990, 0.985, 0.980]
accruals = [0.5, 0.5, 0.5, 0.5]
```

3. Calculate and Show the Fair Fixed Rate

```python
fair_rate = fair_fixed_rate(dfs, accruals)
print("Fair Fixed Rate:", round(fair_rate, 6))
```

4. Visualize Fair Fixed Rates vs. Discount Factors

```python
import matplotlib.pyplot as plt
import numpy as np

base_accruals = [0.5, 0.5, 0.5, 0.5]
discount_factors = np.linspace(0.96, 1.0, 20)

fair_rates = []
for df0 in discount_factors:
    dfs = [df0 - i * 0.005 for i in range(4)]
    fair_rate = fair_fixed_rate(dfs, base_accruals)
    fair_rates.append(fair_rate)
    
plt.figure(figsize=(7,4))
plt.plot(discount_factors, fair_rates, marker='o', label='Fair Fixed Rate')
plt.xlabel('Initial Discount Factor')
plt.ylabel('Fair Fixed Rate')
plt.title('Fair Fixed Rate vs. Initial Discount Factor')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
